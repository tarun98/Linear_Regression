```python
##imports
import numpy as np

from sklearn.datasets import load_boston
```

```python
##get data from sklearn datasets
boston=load_boston()

X=boston.data
Y=boston.target
##adding additional "variable" to account for error/intercept
X=np.transpose(X)
x=np.vstack([X,np.ones(506)])
X=np.transpose(x)

##Seperating first 70% data as train and rest as test
TrX=X[:int(len(X)*(7/10))]
TrY=Y[:int(len(X)*(7/10))]

TeX=X[int(len(X)*(7/10)):]
TeY=Y[int(len(X)*(7/10)):]

##Declaring initial thetas to zero
theta=np.zeros(X.shape[1])
```

```python
###Error FUNCTION


def errorr(x,y,theta):
    return sum((y-np.sum(x*theta,1))**2)/len(y)
    
def GradientStep(X,Y,theta,alpha):
    G=np.zeros(len(theta))
    for j in range(len(theta)):
        for i in range(len(Y)):
            G[j]+=(np.sum(X[i]*theta)-Y[i])*X[i][j]
        
    theta=theta-alpha*G*(1/len(Y))
    return theta
```

```python
##training the model using TrX and TrY to get theta values and minimize error
def GradientDescent(X,Y,theta,alpha,num_iter):
    for i in range(num_iter):
        print("mean squared error in iteration - ",i+1,"is:",errorr(X,Y,theta))
        theta=GradientStep(X,Y,theta,alpha)
    return theta
theta=GradientDescent(TrX,TrY,theta,0.000001,1000)
```

```python
###Using the trained theta value to predict values for Test data and calculate error
print("Real Value | Prediction value | Difference(abs)")
for i in range(len(TeY)):
    pred=sum(theta*TeX[i])
    print("%10.2f | %16.2f | %14.2f" %(TeY[i],pred,abs(TeY[i]-pred)))
print(" ")

```

```python
print("Final Mean squared error of test data is: ",errorr(TeX,TeY,theta))
```
